### Data Visualization Example - Line Graphs with Categorical Data

1. Import libraries

```python
import matplotlib.pyplot as plt
```

2. Create values for x and y points and store them as a list

```python
teams = ['Giants', 'Jets', 'Ravens', 'Rams', 'Bears', 'Eagles']
points = [50, 30, 45, 22, 8, 35]
```

3. Create line graph

```python
plt.title('Football Teams vs Points')
plt.xlabel('Teams')
plt.ylabel('Points')
plt.plot(teams, points)
plt.show()
```

3a. Rotate the x labels (useful with labels with longer names)

```python
plt.title('Football Teams vs Points')
plt.xlabel('Teams')
plt.xticks(rotation = 90)   # note the use of the rotation parameter which is set to 90 degrees
plt.ylabel('Points')
plt.plot(teams, points)
plt.show()
```
