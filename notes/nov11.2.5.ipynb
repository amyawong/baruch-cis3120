### Data Visualization Example - Plotting Multiple Lines on a Single Graph

1. Import libraries

```python
import matplotlib.pyplot as plt
```

2. Create values for x and y points and store them as a list

```python
dates = ['1/7/22', '1/14/22', '1/21/22', '1/28/22', '2/4/22']
opening_price = [40, 20, 45, 48, 30]
closing_price = [38, 10, 45, 50, 41]
```

3. Create line graph

```python
plt.title('Opening vs Closing Stock Prices for a Company')

# various plots have same x values but different y values
plt.plot(dates, opening_price, color = 'b', label = 'opening price', linewidth = 3)
plt.plot(dates, closing_price, color = 'r', label = 'closing price', linewidth = 3)

# different value passed in for loc changes the location of the legend
plt.legend(loc = 4)

plt.show()
```

4. Add a grid to the background of the graph (use a new import)

```python
from matplotlib import style
```

4a. Set the parameter as a grid plot

```python
style.use('ggplot')
```

4b. Create the graph

```python
plt.title('Opening vs Closing Stock Prices for a Company')

# various plots have same x values but different y values
plt.plot(dates, opening_price, color = 'b', label = 'opening price', linewidth = 3)
plt.plot(dates, closing_price, color = 'r', label = 'closing price', linewidth = 3)

# different value passed in for loc changes the location of the legend
plt.legend(loc = 4)

# give the plot a grid-like background
plt.grid(True, color = 'c') # makes it so that grid is in background and is cyan
plt.show()
```
