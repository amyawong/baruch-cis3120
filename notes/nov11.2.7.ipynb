### Data Visualization Example - Bar Graphs

1. Import libraries

```python
import matplotlib.pyplot as plt
```

2. Create values for x and y points and store them as a list

```python
teams = ['Giants', 'Jets', 'Ravens', 'Rams', 'Bears', 'Eagles']
points = [50, 30, 45, 22, 8, 35]
```

3. Create the bar graph

```python
plt.bar(teams, points)
plt.title('Teams vs Points', fontsize = 17)
plt.xlabel('Teams', fontsize = 17)
plt.ylabel('Points', fontsize = 17)
plt.show()
```

4. Create a horizontal bar graph

```python
plt.barh(teams, points)   # only thing that changes is .bar to .barh
plt.title('Teams vs Points', fontsize = 17)
plt.xlabel('Teams', fontsize = 17)
plt.ylabel('Points', fontsize = 17)
plt.show()
```

5. Change the widths of the plot

```python
plt.bar(teams, points, width = 1)   # the width parameter value determines the width
plt.title('Teams vs Points', fontsize = 17)
plt.xlabel('Teams', fontsize = 17)
plt.ylabel('Points', fontsize = 17)
plt.show()
```

6. Change the color of the bars

```python
plt.bar(teams, points, color = 'blue')   # the color parameter value determines the color
plt.title('Teams vs Points', fontsize = 17)
plt.xlabel('Teams', fontsize = 17)
plt.ylabel('Points', fontsize = 17)
plt.show()
```

7. Give each bar an individual color

```python
colors = ['k', 'g', 'r', 'y', 'b', 'w']

plt.bar(teams, points, color = colors)   # colors list gets passed in for color value of bars
plt.title('Teams vs Points', fontsize = 17)
plt.xlabel('Teams', fontsize = 17)
plt.ylabel('Points', fontsize = 17)
plt.figure(figsize = (16, 9)) # change size of plots
plt.show()
```

8. Change the edge of colors of the bar margins

```python
plt.bar(teams, points, color = 'blue', edgecolor = 'r', linewidth = 5)
plt.title('Teams vs Points', fontsize = 17)
plt.xlabel('Teams', fontsize = 17)
plt.ylabel('Points', fontsize = 17)
plt.show()
```

9. Change the edge colors and style of the bar margins

```python
plt.bar(teams, points, color = 'yellow', edgecolor = 'r', linewidth = 5, linestyle = '--')
plt.title('Teams vs Points', fontsize = 17)
plt.xlabel('Teams', fontsize = 17)
plt.ylabel('Points', fontsize = 17)
plt.show()
```
