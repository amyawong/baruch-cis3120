### Data Visualization Example - Plotting a Basic Line Plot

1. Import library

```python
import matplotlib.pyplot as plt
```

2. Create values for x and y points and store them as a list

```python
day = [1, 2, 3, 4, 5, 6, 7]
temperature = [32.3, 35, 22, 45, 50, 28, 40]
```

3. Create line plot

```python
plt.plot(day, temperature)  # day as x-axis and temp as y-axis
plt.title('New York Temperature', fontsize = 20)
plt.axis([0, 5, 0, 60])   # set the limits on the different axes in the format (x_min, x_max, y_min, y_max)
plt.xlabel('Day', fontsize = 15)
plt.ylabel('Temperature', fontsize = 15)
plt.rcParams['figure.figsize'] = [15, 6]   # set the size of the figure

plt.show()
```

3a. Change the color and marker of the line

```python
plt.plot(day, temperature, color = 'b', marker = 'x')  # day as x-axis, temp as y-axis, color as blue, marker as circle
plt.title('New York Temperature', fontsize = 20)
plt.axis([0, 5, 0, 60])   # set the limits on the different axes in the format (x_min, x_max, y_min, y_max)
plt.xlabel('Day', fontsize = 15)
plt.ylabel('Temperature', fontsize = 15)
plt.rcParams['figure.figsize'] = [10, 4]   # set the size of the figure

plt.show()
```

3b. Remove line from line graph and only show plots by changing from line graph to scatter graph

```python
plt.scatter(day, temperature, color = 'r')  # day as x-axis, temp as y-axis, color as red
plt.title('New York Temperature', fontsize = 20)
plt.axis([0, 5, 0, 60])   # set the limits on the different axes in the format (x_min, x_max, y_min, y_max)
plt.xlabel('Day', fontsize = 15)
plt.ylabel('Temperature', fontsize = 15)
plt.rcParams['figure.figsize'] = [10, 4]   # set the size of the figure

plt.show()
```

3c. Change the line style to dashes

```python
plt.plot(day, temperature, color = 'y', marker = '*', linestyle = '--')  # day as x-axis, temp as y-axis, color as yellow, marker as *, line as --
plt.title('New York Temperature', fontsize = 20)
plt.axis([0, 5, 0, 60])   # set the limits on the different axes in the format (x_min, x_max, y_min, y_max)
plt.xlabel('Day', fontsize = 15)
plt.ylabel('Temperature', fontsize = 15)
plt.rcParams['figure.figsize'] = [10, 4]   # set the size of the figure

plt.show()
```

3d. Change the line width

```python
plt.plot(day, temperature, color = 'c', marker = '*', linestyle = '--', linewidth = 3)  # day as x-axis, temp as y-axis, color as cyan, marker as *, line as --, line width as 3
plt.title('New York Temperature', fontsize = 20)
plt.axis([0, 5, 0, 60])   # set the limits on the different axes in the format (x_min, x_max, y_min, y_max)
plt.xlabel('Day', fontsize = 15)
plt.ylabel('Temperature', fontsize = 15)
plt.rcParams['figure.figsize'] = [10, 4]   # set the size of the figure

plt.show()
```

3e. Change the marker size

```python
plt.plot(day, temperature, color = 'c', marker = '*', linestyle = '--', linewidth = 3, markersize = 15)  # day as x-axis, temp as y-axis, color as cyan, marker as *, line as --, line width as 3, marker size as 15
plt.title('New York Temperature', fontsize = 20)
plt.axis([0, 5, 0, 60])   # set the limits on the different axes in the format (x_min, x_max, y_min, y_max)
plt.xlabel('Day', fontsize = 15)
plt.ylabel('Temperature', fontsize = 15)
plt.rcParams['figure.figsize'] = [10, 4]   # set the size of the figure

plt.show()
```
