3. Authorization URL

4. Make a POST request to API Server  
by sending credentials to server in form of dictionary  
first arg is AUTH_URL, second arg is dictionary  

```python
auth_response = requests.post(AUTH_URL, {
  'grant_type': 'client_credentials',
  'client_id': CLIENT_ID,
  'client_secret': CLIENT_SECRET
})
```

4a. Check status code

```python
auth_response.status_code    # should print 200
```

```python
if auth_response.status_code == 200:

  # 5. Convert the response to JSON
  auth_response_data = auth_response.json()
  print(auth_response_data)   # token changes, once it expires, need to make a new token
```

```python
  # 6. Save the access token
access_token = auth_response_data['access_token']
access_token
```

```python
  # 7. Store access token as a header
headers = {
    'Authorization': f'Bearer {access_token}'
  }
headers
```
